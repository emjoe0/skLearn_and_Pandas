```python
# getting the load_boston datasets for conversion
# Note: sklearn datasets usally comes in dictionary format which has to be converted.
boston = load_boston()
boston
```

```python
len(boston)
```

```python
x = boston.data #the data
y = boston.target #the 
bosCol = boston.feature_names #column names of the dataset
```

```python
# loading the data to a pandas dataframe
df = pd.DataFrame(x)
df
```

```python
# adding the column header to the dataset
df.columns = list(bosCol)
df
```

```python
#adding the class variable as additinal columns to my dataset
df['CLASS'] = list(y)
df
```

```python
# inpecting the boston dataset
df.info()
```

```python
df.loc[df.duplicated() == True] #returns only row which the duplicate status is True
```

```python
# checking for duplicate rows in the entire dataset (including all columns)
duplicatedRows = df[df.duplicated()]
duplicatedRows
```

```python
# checking for duplicate cells in the age feature
duplicatedCol = df[df.duplicated(['AGE'])]
duplicatedCol
```

```python
# checking for duplicate cells in the AGE, TAX & CHAS features
duplicatedCols = df[df.duplicated(['AGE','TAX','CHAS'])]
duplicatedCols
```

```python
checking the total number duplicate samples in AGE, TAX & CHAS features
len(duplicatedCols)
```

```python
# using pandas to extract the samples containing age that is greater than 95 in thane AGE column
extractedRows = df.loc[df['AGE'] > 95]
extractedRows
```

```python
# appending the extracted rows to the original dataset to creates duplicated rows for excercise purpuse
duplicatedRows = pd.concat([df, extractedRows])
duplicatedRows
```

```python
# getting only the newly added duplicate rows in the original dataset
duplicatedRows2 = duplicatedRows[duplicatedRows.duplicated()]
duplicatedRows2
```

```python
#Removing duplicated rows but keeping the first row. I can as well keep the last row or remove all duplicates entirely 
# subset = the column names I want the duplicated rows to be removed
df_noDuplicates = duplicatedRows.drop_duplicates(keep='first')
df_noDuplicates
```

```python
# saving the dataframe without duplicated rows to the orginal dataframe using inplace=True
duplicatedRows.drop_duplicates(keep='first', inplace=True)
duplicatedRows
```
